#Pie Chart
representing the unique values of the category_1 column  and there percentage in category_1

```python
#pie Chart
import random
fig_1=plt.figure(figsize=(8,5))
axes_1=fig_1.add_axes([0.1,0.1,0.9,0.9])
Category_1_list=['Sports, Books and More','Electronics',"Men's wear","Women's wear",
 'Bady and Kids','Home and Furniture']
values=[]
for i in df["category_1"].unique():
     values.append((df["category_1"]==i).sum())
colors=[]
for j in range(6):
     rgb=(random.uniform(0,.5,),random.uniform(0,.5),random.uniform(0,0.5))
     colors.append(rgb)
explode=[0]*6
explode[0]=0.2

wedges, texts, autotexts = plt.pie(values, explode=explode, labels=Category_1_list, colors=colors, 
                                   autopct='%1.0f%%', shadow=True, startangle=140, 
                                   textprops=dict(color="w"))
plt.legend(wedges, Category_1_list, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("share of each Product in Category_1")
```

#Representing rating
taking ten sample ratings from the product_rating column and there count and representing graphycally.

```python
#Rating
fig_r=plt.figure(figsize=(8,5))
axes_r=fig_r.add_axes([0.1,0.1,0.9,0.9])
rating=pd.Series(df["product_rating"].unique()).sample(n=10,replace=False,random_state=1)
# rating=df["product_rating"].sort_values(ascending=False)

values_r=[]
for i in rating:
    values_r.append((df["product_rating"]==i).sum())
print(rating)
print(values_r)
plt.bar(rating.values,values_r,width=0.5)
plt.xlabel("product_rating")
plt.ylabel("product_rating count")
plt.title("product_rating VS product_rating count")
```
